```python
import os
import re
import json
from pathlib import Path
from concurrent.futures import ProcessPoolExecutor, as_completed

import pandas as pd
from tqdm import tqdm

OncDRS_PATH = Path('/data/gusev/PROFILE/CLINICAL/OncDRS/')

STAGE_PATTERNS = [
    r'\bstage\s+(?:IV|III|II|I)[ABCabc]?\b',   # Stage IIIA, stage IV
    r'\bstage\s+[0-4][ABCabc]?\b',             # stage 4, stage 2b
    r'\bstage\s+(?:one|two|three|four)\b',     # spelled out
]
stage_re = re.compile('|'.join(STAGE_PATTERNS), re.IGNORECASE)

# Precompiled cleanup patterns (module-level so each worker process reuses them).
_WS_RUN = re.compile(r'[ \t]+')
_BLANK_RUN = re.compile(r'\n{3,}')
# Strip C0/C1 control chars but keep \n. Much faster than a per-char isprintable() loop.
_CTRL = re.compile(r'[\x00-\x08\x0b\x0c\x0e-\x1f\x7f-\x9f]')


def clean_note(text):
    text = text.replace('\r\n', '\n').replace('\r', '\n')
    text = _WS_RUN.sub(' ', text)
    text = _BLANK_RUN.sub('\n\n', text)
    text = _CTRL.sub('', text)
    return text.strip()


def load_json_file(filepath):
    with open(filepath, 'r') as f:
        return json.load(f)['response']['docs']


def process_one_file(args):
    """Worker: load a single JSON file and return only the staged notes."""
    year, filename = args
    filepath = OncDRS_PATH / f'CLINICAL_TEXTS_{year}' / filename
    rows = []
    for doc in load_json_file(filepath):
        rpt_txt = doc.get('RPT_TEXT')
        if not rpt_txt:
            continue
        # Search first, clean only on a hit — cleaning is the expensive part.
        if stage_re.search(rpt_txt):
            rows.append((
                doc.get('DFCI_MRN'),
                doc.get('EVENT_DATE'),
                doc.get('RPT_TYPE'),   # None where the schema omits it (e.g. discharge summaries)
                clean_note(rpt_txt),
            ))
    return rows


def build_file_list():
    years = ['2024_03', '2025_03', '2025_11']
    year_file_dict = {
        year: [
            f for f in os.listdir(OncDRS_PATH / f'CLINICAL_TEXTS_{year}')
            if f.endswith('.json')
        ]
        for year in years
    }

    nested = []
    base_2026 = OncDRS_PATH / 'CLINICAL_TEXTS_2026_03'
    for subdir in os.listdir(base_2026):
        for file in os.listdir(base_2026 / subdir):
            if file.endswith('.json'):
                nested.append(f'{subdir}/{file}')
    year_file_dict['2026_03'] = nested

    return [(year, fn) for year, files in year_file_dict.items() for fn in files]


def n_workers(cap=8):
    """Cores actually allocated to this job, not the whole node.

    os.cpu_count() reports every core on the physical node and ignores the
    SLURM cgroup, which leads to oversubscription and OOM-killed workers.
    sched_getaffinity respects the allocation. Capped to keep peak memory
    sane, since each worker holds a full JSON file at once.
    """
    try:
        allocated = len(os.sched_getaffinity(0))
    except AttributeError:        # not Linux
        allocated = os.cpu_count() or 1
    return max(1, min(cap, allocated))


def main():
    unpacked_files = build_file_list()

    all_rows = []
    files_with_hits = 0
    workers = n_workers()
    with ProcessPoolExecutor(max_workers=workers) as ex:
        futures = {ex.submit(process_one_file, item): item for item in unpacked_files}
        bar = tqdm(
            as_completed(futures),
            total=len(futures),
            desc=f'Scanning notes ({workers} workers)',
            unit='file',
        )
        for i, fut in enumerate(bar, start=1):
            year, filename = futures[fut]
            try:
                rows = fut.result()
            except Exception as e:
                raise RuntimeError(f'Failed on {year}/{filename}: {e!r}') from e
            all_rows.extend(rows)
            if rows:
                files_with_hits += 1
            bar.set_postfix(
                notes=len(all_rows),
                hit_rate=f'{files_with_hits / i:.1%}',
                refresh=False,
            )

    full_pull = pd.DataFrame(
        all_rows,
        columns=['DFCI_MRN', 'EVENT_DATE', 'RPT_TYPE', 'RPT_TEXT'],
    )
    return full_pull

if __name__ == '__main__':
    full_pull = main()
    print(full_pull.shape)
```
